````python
#OBJECT ORIENTED PROGRAMMING

#Classes or type- blueprints for their instances
#objects- instances of classes with their own attributes and methods which can be accesed through . (dot) operator
#hasattr(class,attr)  getattr(class,attr)
#special methods- to define an objects high level behaviour __add__()for defining + addition operation   __getitem__()for defining obj[key]  __repr__()  __pow__()  __mul__()  __setitem__()  __iter__()for defining iteration opertaion
#instance methods- defined within class, instance passed as first argument
#class methods- class object passed as first argument @classmethod  
#static methods- all arguments must be passed by user @static method
#self keyword stresses that we want to run this funtion for *this* specific object

class MyGuy:
    #class attributes
    x = 1.0
    y = 'hi'

#__init__ is automatically called while instance creation "under the hood"
    def __init__(self,name):
        #object attribute and not accesible by class
        self.name = name
        self.obj_attr = 'Pikachu '+ name

    def fly(self):
        return 'i am flying'

obj1 = MyGuy('Takashi')

#all correct
print(MyGuy.x )
print(obj1.x )
print(MyGuy.fly(obj1))

#all wrong
print(MyGuy.fly())
print(MyGuy.obj_attr)
````

````python
#INHERITANCE

#issubclass(class,class)

class MyKid(MyGuy):
    def __init__(self,name,age):
        super().__init__(self,name)
        self.age = age
````
